```python
import pandas as pd
import ast  # For safely converting string to dictionary
import json

# Replace NaN with an empty dictionary string '{}'
df['amenities'] = df['amenities'].fillna('{}')

# Convert 'true'/'false' to 'True'/'False' for Python compatibility
df['amenities'] = df['amenities'].str.replace('true', 'True').str.replace('false', 'False')

# Safely convert the string to a dictionary
df['amenities'] = df['amenities'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else {})

# Expand dictionary into separate columns while keeping other columns unchanged
amenities_df = df['amenities'].apply(pd.Series)  # Convert dict to individual columns
df = df.drop(columns=['amenities']).join(amenities_df)  # Merge new columns back into original DataFrame

# Display updated DataFrame
print(df)
```

```python
df
```

```python
#Removing Empty Columns
df.drop(columns=['combineDescription', 'weight'], inplace=True)
```

```python
#Replacing Blank Values with 0 
df['balconies'].fillna(0, inplace=True)
df['maintenanceAmount'].fillna(0, inplace=True)
```

```python
#Removinging Duplicate Records 
df.drop_duplicates(inplace=True)
print(df)
```

```python
#Identifying Numerical Columns 
num_cols = df.select_dtypes(include=['number'])
```

```python
from scipy.stats import zscore

# Compute Z-score for numerical columns
num_cols_zscore = num_cols.apply(zscore)

# Display the first few rows
print(num_cols_zscore.head())
```

```python
# identify outliers based on z-score threshold
threshold = 3
outlier_rows, outlier_cols = np.where(num_cols_zscore > threshold)
```

```python
#Removing Outliers
df_clean = df.drop(df.index[outlier_rows])
```

```python
#Storing Categorical Columns in a variable
cat_cols = df.select_dtypes(include=['object', 'category']).columns

print("Categorical columns:", cat_cols.tolist())
```

```python
#Encoding Categorical variables
for col in cat_cols:
    df[col] = df[col].astype('category').cat.codes
```

```python
#Seperate Features and Target
X = df.drop([ 'shortUrl','rent_amount','reactivationSource','propertyType','parkingDesc','ownerName','localityId','propertyTitle','id','facingDesc'], axis=1)
y = df["rent_amount"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
```

```python
# create RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
import sklearn.metrics
from sklearn.metrics import mean_squared_error
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split, cross_val_score

# Initialize the model
model = RandomForestRegressor(random_state=42)

# Define the range of features to test
n_features_range = range(1, X.shape[1] + 1)  # From 1 to total number of features
rmse_scores = []

# Loop through different numbers of features
for n in n_features_range:
    # Initialize RFE
    rfe = RFE(estimator=model, n_features_to_select=n)
    
    # Transform the data to select the top n features
    X_rfe = rfe.fit_transform(X, y)
    
    # Perform cross-validation
    scores = cross_val_score(model, X_rfe, y, scoring='neg_root_mean_squared_error', cv=5)
    
    # Store the average RMSE score
    rmse_scores.append(-np.mean(scores))

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(n_features_range, rmse_scores, marker='o')
plt.xlabel('Number of Features')
plt.ylabel('Cross-Validated RMSE')
plt.title('Optimal Number of Features')
plt.grid(True)
plt.show()

# Find the optimal number of features
optimal_n_features = n_features_range[np.argmin(rmse_scores)]
print(f"Optimal number of features: {optimal_n_features}")
```

```python
# Initialize RFE with the optimal number of features
rfe = RFE(estimator=model, n_features_to_select=optimal_n_features)

# Fit RFE and transform the data
X_rfe = rfe.fit_transform(X, y)

# Get the names of the selected features
selected_features = X.columns[rfe.support_]
print("Selected features:", selected_features)
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_rfe, y, test_size=0.2, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Assuming y_test and y_pred are defined
# y_test: True target values
# y_pred: Predicted target values

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))  # Manual calculation
print("Test RMSE:", rmse)

# Calculate R² score
r2 = r2_score(y_test, y_pred)
print("Test R²:", r2)
```

```python
# Initialize the Gradient Boosting model
gb_model = GradientBoostingRegressor(random_state=42)

# Train the model
gb_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = gb_model.predict(X_test)

# Evaluate the model
# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))  # Manual calculation
print("Test RMSE:", rmse)

# Calculate R² score
r2 = r2_score(y_test, y_pred)
print("Test R²:", r2)
```

```python
from sklearn.linear_model import LinearRegression
# Define base models using the pre-trained models
base_models = [
    ('random_forest', model),  # Use the pre-trained Random Forest model
    ('gradient_boosting', gb_model)  # Use the pre-trained Gradient Boosting model
]

# Define meta-model
meta_model = LinearRegression()

# Create the stacking ensemble
stacking_model = StackingRegressor(
    estimators=base_models,
    final_estimator=meta_model,
    cv=5  # Cross-validation folds for training the meta-model
)

# Train the stacking model
stacking_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = stacking_model.predict(X_test)
```

```python
# Evaluate the model
# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))  # Manual calculation
print("Test RMSE:", rmse)

# Calculate R² score
r2 = r2_score(y_test, y_pred)
print("Test R²:", r2)
```
